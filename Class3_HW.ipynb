```
import pandas as pd
import spacy
```

```
data = pd.read_json('http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz',lines=True)
df = pd.DataFrame(data)
```

```
spacyy = spacy.load('en_core_web_sm')
```

```
text = df.reviewText[4]
```

```
sent = spacyy(text)
```

```
def SpaPipe(xyz):
  sentence = spacyy(xyz)
  Noun = []
  Token = []
  Lemm = []
  Tags = []
  Dp = []
  for i in sentence:
    Token.append(i.text)
    Lemm.append(i.lemma_)
    Tags.append(i.pos_)
    Dp.append(i.dep_)
  for n in sentence:
    if n.pos_ == 'NOUN':
      Noun.append(n.text)
  final = {'Tokens' : Token,'Lemma' : Lemm,'Tags' : Tags, 'Dependency' : Dp,'Nouns' : Noun}
  return final
```

```
SpaPipe(text)
```
